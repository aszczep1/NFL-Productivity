### Player Stats 
<br><i>Note: From <a href='https://www.kaggle.com/datasets/kendallgillies/nflstatistics'>Kaggle.com</a></i>

```python
#Import the datasets 
qb = pd.read_csv(r'Game_Logs_Quarterback.csv')
rb = pd.read_csv(r'Game_Logs_Runningback.csv') 
wr = pd.read_csv(r'Game_Logs_Wide_Receiver_and_Tight_End.csv') 
```

```python
#Combine the datasets for one large one containing all skill position players
stats = pd.concat([qb, rb, wr], ignore_index = True )
```

```python
#See the columns in the dataset 
stats.columns
```

```python
qb.columns
```

```python
stats.fillna(value = 0, inplace = True )
qb.fillna(value = 0, inplace = True)
```

```python
#select the columns we are going to use for each 
stats = stats[['Player Id', 'Name', 'Year', 'Season', 'Home or Away', 'Week','Game Date', 'Opponent', 'Outcome', 'Passing Yards', 'TD Passes', 'Rushing Yards', 'Rushing TDs', 'Receiving Yards', 'Receiving TDs' ]]
qb = qb[['Player Id', 'Name', 'Year', 'Season', 'Home or Away','Week','Game Date', 'Opponent', 'Outcome', 'Passing Yards', 'TD Passes', 'Rushing Yards', 'Rushing TDs', 'Ints', 'Passer Rating' ]]
```

```python
tot_cols_stats = ['Rushing Yards', 'Receiving Yards', 'Passing Yards','Rushing TDs', 'Receiving TDs', 'TD Passes']
tot_cols_qb = ['Rushing Yards', 'Passing Yards','Rushing TDs', 'TD Passes'] #target the columns to turn to numerics 
for col in tot_cols_stats: 
    stats[col] = pd.to_numeric(stats[col], errors = 'coerce') #turn the columns to numerics 
for col in tot_cols_qb: 
    qb[col] = pd.to_numeric(qb[col], errors = 'coerce') 
total_yards_cols_st = stats[['Rushing Yards', 'Receiving Yards', 'Passing Yards']]
total_td_st = stats[['Rushing TDs', 'Receiving TDs', 'TD Passes']]
total_yards_cols_qb = qb[['Rushing Yards','Passing Yards']] #target the columns that we would like to add togther
total_td_qb = qb[['Rushing TDs', 'TD Passes']]
tot_yards_stat = total_yards_cols_st.sum(axis = 1) #add together for total yards
tot_td_stat = total_td_st.sum(axis = 1) #add together for total touchdowns 
tot_yards_qb = total_yards_cols_qb.sum(axis = 1) #add together for total yards
tot_td_qb = total_td_qb.sum(axis = 1) #add together for total touchdowns
```

```python
stats['Total Yards'] = tot_yards_stat #assign the lists to the columns 
stats['Total TDs'] = tot_td_stat
qb['Total Yards'] = tot_yards_qb #assign the lists to the columns 
qb['Total TDs'] = tot_td_qb
```

```python
stats = stats.drop(total_yards_cols_st + total_td_st, axis = 1) #drop all the columns in these lists
qb = qb.drop(total_yards_cols_qb + total_td_qb, axis = 1) #drop all the columns in these lists
```

```python
qb = qb[qb["Year"] >= 2014] #only take rows with years 2014 and after
qb = qb[qb["Season"] != "Preseason"] #only take rows for regular season and postseason 
qb = qb[qb["Season"] != "Pro Bowl"]
qb['Season'] = qb['Season'].replace({'Regular Season':0, 'Postseason':1}) #create dummy varibales 
qb['Home or Away'] = qb['Home or Away'].replace({'Away':0, 'Home':1})
qb['Outcome'] = qb['Outcome'].astype('category')
#qb.isnull().sum() #check for nulls in the data
```

```python
stats = stats[stats["Year"] >= 2014] #only take rows with years 2014 and after
stats = stats[stats["Season"] != "Preseason"] #only take rows for regular season and postseason 
stats = stats[stats["Season"] != "Pro Bowl"]
stats['Season'] = stats['Season'].replace({'Regular Season':0, 'Postseason':1}) #create dummy varibales 
stats['Home or Away'] = stats['Home or Away'].replace({'Away':0, 'Home':1})
stats['Outcome'] = stats['Outcome'].astype('category')
#stats.isnull().sum() #check for nulls in the data
```

```python
games.isnull().any()
games = games.drop('Unnamed: 0', axis = 1)
```

```python
games
```

```python
#make sure that the merge columns are the same type otherwise an error will occur 
games['Week'] = pd.to_numeric(games['Week'], errors = 'coerce')
stats['Week'] = pd.to_numeric(stats['Week'], errors = 'coerce')

print(type(games['Week'][3]))
print(type(stats_2['Week'][7]))
print(type(qb['Week'][12]))
print(type(games['Game Date'][3]))
print(type(stats_2['Game Date'][7]))
```

```python
stats_3 = stats_2.merge(games, on = ["Year",  'Opponent', 'Week'], how = 'left')
```

```python
teams_index = stats_3['Team'].index[0:]
team_home = []
for i in teams_index:
    if stats_3['Home or Away'][i] == 1: 
        team_home.append(stats_3['Team'][i])
    else:
        team_home.append(stats_3['Opponent'][i])
stats_3['Home Team'] = team_home
```

```python
stats_3 = stats_3.drop('Game Date_y', axis = 1)
```

```python
#convert time data into categorical variable for primetime
print(type(stats_3["Time"][0]))
```

```python
stat_index = stats_3.index[0:]
primetime = []
for num in stat_index:
    if stats_3["Season"][num] == 0.0:
        if stats_3['Time'][num].startswith(('1','3','4','6')):
            primetime.append(0)
        else:
            primetime.append(1)
    else: 
        primetime.append(1)
stats_3['Primetime'] = primetime
stats_3 
```
