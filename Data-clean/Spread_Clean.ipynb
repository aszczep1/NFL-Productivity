```python
import pandas as pd
import numpy as np
```

```python
spread = pd.read_csv(r'spreadspoke_scores.csv')
```

```python
spread.head()
```

```python
spread = spread.rename(columns = {'schedule_season':'Year', 'schedule_week':'Week', 'schedule_date':'Game Date', 'team_home':'Home Team','team_away':'Away Team', 'schedule_playoff':'Season'})
```

```python
spread_drop = [ 'weather_detail', 'score_home', 'score_away', 'spread_favorite', 'over_under_line', 'stadium', 'stadium_neutral', 'team_favorite_id', 'weather_humidity']
```

```python
spread = spread.drop(spread_drop, axis=1)
```

```python
spread = spread[spread['Year'] >= 2014]
spread = spread[spread['Year'] <= 2016]
```

```python
spread.isnull().sum()
```

```python
spread['Week'] = spread['Week'].replace({'Wildcard': 18, 'Division':19, 'Conference':20, 'Superbowl':22})
spread['Season'] = spread['Season'].replace({False: 0, True:1})
```

```python
spread
```

```python
spread.to_csv('spread_cleaned.csv')
```
